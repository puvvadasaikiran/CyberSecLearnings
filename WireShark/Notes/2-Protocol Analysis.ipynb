#DTP(Dynamic Trunking Protocol)

image.png

> A trunk is a communications line or link designed to carry multiple signals simultaneously to provide network access between two point

> Trunks typically connect switching centers in a communications system. 

> The signals can convey any type of communications data

> DTP is between Switches

> By default Cisco Switches are set to Dynamic Auto

> if two switches have **Dynamic Auto** setting None of them will initiate trunk

> One of the side should be set to **Dynamic Desirable** for a connection to be established

> and when a switch intiates the connection, both negotiate to use common protocol (ISL and 802.1Q)



#CDP(Cisco Discovery Protocol) and LLDP(Link Layer Discovery Protocol)

> Operates at Layer 2 of OSI Model

> CDP only discovers Cisco devices and not every device

> Cisco devices should enable LLDP to be discoverable by other Manufacturers

> these protocols are helpful to know the devices in the network 

> CDP is not dependent on IP

> CDP can discover only directly connected neighbours



#Open Shortest Path First(OSPF) 

> Both routers need to share password to form a connection and share the knowledge databases

> They need to use same network masks to form a relation

>They can't for a connection if hello Interval is different or Dead Interval is different


#EIGRP(Enhanced Interior Gateway Routing Protocol)

> It is an advanced distance-vector routing protocol that is used on a computer network for automating routing decisions and configuration.

> it uses MD5 by default to go on with the connection

> EIGRP K-Values and AS(Autonomous System) Number should match for connection 


#BGP (Border Gateway Protocol)

> gateway protocol that enables the internet to exchange routing information between autonomous systems (AS)

> Internet is heavily dependent on BGP

> BGP selects the routes

> Two types of BGP => Internal BGP (iBGP) and External BGP (eBGP)

> iBGP used between inter-BGP connections eBGP for ISP-BGP connections
