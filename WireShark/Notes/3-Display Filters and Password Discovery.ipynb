#Display Filters
Display filters allow us to filter the wireshark packet data

Eg: We can filter out by source and destination IP as well as the protocols (TCP,UDP etc) and port numbers

##Routers re-write the mac address
lets say we have ping PC2 from PC1 via Router1

the ping request from PC1-> PC2 will have srcMAC=PC1MAC and dstMAC=Router1MAC

when this goes to the router, PC1 Mac will be replaced with Router MAC address ie. srcMAC=Router1MAC and dstMAC=PC2MAC



```
Commands used:
ip.src==10.1.1.1 or ip.src == 10.1.2.1
dns //filters out dns requests
arp //shows ARP requests

//Just C lang type Syntaxing like &&,|| for and,or
```



## Wireshark has two types of filtering Display Filter and Capture Filter
Capture Filters are used to just capture some traffic and not everything that goes through network



```
Goto Capture tab > Options > "just write filters below"

eg: tcp port 23 > captures tcp traffic that goes from port 23

Advantage of Capturing Filter is You are Focussed on particular traffic and can reduce the capture file size  
```



##While using Network Protocols we need to be aware of protocols that use http (coz http uses Clear Text Transfering) and these protocols can be snooped to get login information easily
